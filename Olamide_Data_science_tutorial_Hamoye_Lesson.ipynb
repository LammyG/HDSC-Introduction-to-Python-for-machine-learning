```python
# groupby the fuel type code year and print the first entries in all the groups formed
fuel_data.groupby('fuel_type_code_pudl').first()
```

```python
# split the fuel data into two and reset the index
fuel_df1 = fuel_data.iloc[0:19000].reset_index(drop=True)
fuel_df2 = fuel_data.iloc[19000:].reset_index(drop=True)
fuel_df1
```

```python
fuel_df2
```

```python
# check that the sum of both dataframes sum to the expected length
assert len(fuel_data) == (len(fuel_df1)) + (len(fuel_df2)), "They are not the same"
print("They match and are equal")
```

```python
# an inner merge will loose rows that do not match in both dataframes
pd.merge(fuel_df1, fuel_df2, how="inner")
```

```python
# outer merge will return rows that do not merge in both dataframes
pd.merge(fuel_df1, fuel_df2, how="outer")
```

```python
# remove rows from the right dataframe that do not have a match with the left dataframe
# and keeps all rows from the left
pd.merge(fuel_df1, fuel_df2, how="left")
```

```python
# concatenate and keep their respective index
pd.concat([fuel_df1, fuel_df2]).reset_index(drop=True)
```
